```python
# Simplified formula for converting Fahrenheit to Celsius

import numpy as np
import matplotlib.pyplot as plt

(98.0 - 32) * 5 / 9

offset_exact = 32
offset_approx = 30  # cross at ~50F
# offset_approx = 28  # cross at ~68F

fexact = lambda f: (f - offset_exact) * 5 / 9
fapprox = lambda f: (f - offset_approx) * 5 / 10

# f1, f2 = 32, 212
f1, f2 = 32, 80

plt.figure()
plt.plot([f1, f2], [fexact(f1), fexact(f2)], label=f'Exact:   (F - {offset_exact}) * 5/9')
plt.plot([f1, f2], [fapprox(f1), fapprox(f2)], label=f'Approx: (F - {offset_approx}) / 2')
plt.xlabel('Fahrenheit')
plt.ylabel('Celsius')
plt.legend()
plt.grid()
```
